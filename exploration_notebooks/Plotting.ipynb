```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
numbers_rdf_to_text = [0.139747664451507, 0.212847924491958, 0.226716748531253, 0.23509582972166, 0.238755658287585, 0.239044592121737]
numbers_question_answering =[0.093469910371319, 0.415012804097311, 0.488556338028169, 0.506001920614597, 0.512003841229193, 0.516165172855314]
steps = [1000, 2000, 3000, 4000, 5000, 6000]
items = [step * 512 for step in steps]
```

```python
f, ax = plt.subplots(1,figsize=(12,5))
ax.plot(items, numbers_question_answering)
ax.set_ylim(ymin=0)
ax.set_ylim(ymax=0.8)
ax.set_xlabel('Sentences', fontsize=12)
ax.set_ylabel('Accuracy')
f.suptitle('QA Accuracy @1 over GOOAQ during training', fontsize=16)
plt.show(f)
```

```python
f, ax = plt.subplots(1,figsize=(12,5))
ax.plot(items, numbers_rdf_to_text)
ax.set_ylim(ymin=0)
ax.set_ylim(ymax=.5)
ax.set_xlabel('Sentences', fontsize=12)
ax.set_ylabel('Accuracy', fontsize=12)
f.suptitle('RDF-to-text Accuracy @1 over WebNLG-wikidata during training', fontsize=16)
plt.show(f)
```
